```python
from sympy import *
init_printing()
import matplotlib.pyplot as plt
```

```python
def radians(i):
    return i / 360 * 2 * pi.n(8)
```

```python
h = 0.016
alpha = 24
delta = 330
rho = 0.21
omega_1 = 12.566
```

```python
phi_p = delta / Rational(7, 4)
v_1, v_2 = symbols('v_1, v_2')
phi = symbols('varphi')
a1 = radians(phi_p / 4) * v_1 + v_1 * radians(phi_p / 2)
```

```python
a1
```

```python
a2 = radians(phi_p / 8) * v_2 + v_2 * radians(phi_p / 4)
```

```python
a2
```

```python
s1 = solve([a1 - a2, a1 - h], [v_1, v_2])
```

```python
s1
```

```python
var1 = [0, (phi_p / 4).n(5), (3 * phi_p / 4).n(5), phi_p, 1.25 * phi_p.n(5), 1.375 * phi_p.n(5), 1.625 * phi_p.n(5), delta]
var1 = [radians(i) for i in var1]
var2 = [0, s1[v_1], s1[v_1], 0, 0, -s1[v_2], -s1[v_2], 0]


fig_01 = plt.figure(figsize = [9.00, 7.00])
plt.plot(var1, var2, linewidth = 2, color = 'black')
plt.grid(True)
plt.show()
```

```python
v1, v2 = s1[v_1], s1[v_2]
v1, v2
```

```python
var1 = [0, (phi_p / 4).n(5), (phi_p / 4).n(5) + 0.000001, (3 * phi_p / 4).n(5), (3 * phi_p / 4).n(5) + 0.000001,
        phi_p, phi_p + 0.000001, 1.25 * phi_p.n(5), 1.25 * phi_p.n(5) + 0.000001, 1.375 * phi_p.n(5), 1.375 * phi_p.n(5) + 0.000001, 1.625 * phi_p.n(5),
        1.625 * phi_p.n(5) + 0.000001, delta]
var1 = [radians(i) for i in var1]
var2 = [v1 / radians(phi_p / 4), v1 / radians(phi_p / 4), 0, 0, -v1 / radians(phi_p / 4),
        -v1 / radians(phi_p / 4), 0, 0, -v2 / radians(phi_p / 8), -v2 / radians(phi_p / 8), 0, 0, v2 / radians(phi_p / 8),
       v2 / radians(phi_p / 8)]

fig_01 = plt.figure(figsize = [9.00, 7.00])
plt.plot(var1, var2, linewidth = 2, color = 'black')
plt.grid(True)
plt.show()
```

```python
v = (v1 / radians(phi_p / 4) * phi, v1, -v1 / radians(phi_p / 4) * phi, 0, -v2 / radians(phi_p / 8) * phi, -v2, v2 / radians(phi_p / 8) * phi)


var2 = [integrate(v[0], phi).subs(phi, radians(i)) for i in range(round((phi_p / 4).n(5)))]
last = var2[-1]
#s_dop = solve([v[1].subs(phi, radians(phi_p / 4)) + l1], [l1])
var2 += [0.0001 + last + integrate(v[1], phi).subs(phi, radians(i) - radians(phi_p / 4)) for i in range(round((phi_p / 4).n(5)), round((0.75 * phi_p).n(5)))]
l1 = symbols('l_1')
s_dop = solve([v[2].subs(phi, radians(phi_p)) + l1], [l1])
last = var2[-1]
var2 += [0.00015 + last + integrate(v[2] + s_dop[l1], phi).subs(phi, radians(i)) - integrate(v[2] + s_dop[l1], phi).subs(phi, radians(0.75 * phi_p)) for i in range(round(0.75 * phi_p), round(phi_p))]
last = var2[-1]
var2 += [0 + last for i in range(round((phi_p).n(5)), round((1.25 * phi_p).n(5)))]
last = var2[-1]
lll = last
s_dop = solve([v[4].subs(phi, radians(1.25 * phi_p.n(5))) + l1], [l1])
var2 += [last + integrate(v[4] + s_dop[l1], phi).subs(phi, radians(i)) - integrate(v[4] + s_dop[l1], phi).subs(phi, radians(1.25 * phi_p)) for i in range(round((1.25 * phi_p).n(5)), round((1.375 * phi_p).n(5)))]
s_dop = solve([v[5].subs(phi, radians(1.375 * phi_p) ) + l1], [l1])
last = var2[-1]
var2 += [-0.0002 + last + integrate(v[5], phi).subs(phi, radians(i) - radians(1.375 * phi_p)) for i in range(round((1.375 * phi_p).n(5)), round((1.625 * phi_p).n(5)))]
last = var2[-1]
s_dop = solve([v[6].subs(phi, radians(1.75 * phi_p.n(5))) + l1], [l1])
llll = last
var2 += [-0.0004 + last + integrate(v[6] + s_dop[l1], phi).subs(phi, radians(i)) - integrate(v[6] + s_dop[l1], phi).subs(phi, radians(1.625 * phi_p)) for i in range(round((1.625 * phi_p).n(5)), round((1.75 * phi_p).n(5)))]

var1 = [radians(i) for i in range(len(var2))]

fig_03 = plt.figure(figsize = [9.00, 7.00])
plt.plot(var1, var2, linewidth = 2, color = 'black')
plt.grid(True)
plt.show()
```
